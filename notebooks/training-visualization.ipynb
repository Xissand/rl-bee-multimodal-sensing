```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python

SMALL_SIZE = 18
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
df1 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round1/logs/progress.csv")
df2 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round2/logs/progress.csv")
df3 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round3/logs/progress.csv")
df4 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round4/logs/progress.csv")
df5 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round5/logs/progress.csv")
df6 = pd.read_csv("/Users/teska/phd/classes/neuromatch/project/rl-bee-multimodal-sensing/models/bee-world-500k_v3/round6/logs/progress.csv")
```

```python
fig, ax = plt.subplots(1, 5, figsize=(30, 5), sharey=False)
ax[0].plot(df1['eval/mean_ep_length'].dropna())
ax[1].plot(df2['eval/mean_ep_length'].dropna())
ax[2].plot(df3['eval/mean_ep_length'].dropna())
ax[3].plot(df4['eval/mean_ep_length'].dropna())
ax[4].plot(df5['eval/mean_ep_length'].dropna())

ax[0].set_ylabel('Mean Episode Length')

for n, axis in enumerate(ax):
    axis.set_title(f'Training round: {n+1}')
    axis.set_xlabel('Episode number')
    
plt.savefig("/Users/teska/phd/classes/neuromatch/project/figures/mean_episode_length.png", dpi=300, bbox_inches='tight')
```
