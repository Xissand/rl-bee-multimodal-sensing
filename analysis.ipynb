```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
MAX_STEPS = 5000
```

```python
def count_good(ar):
    return len(ar[ar==MAX_STEPS])/len(ar)
```

```python
def filter_good(ar):
    return ar[ar!=MAX_STEPS]
```

```python
good = np.loadtxt("models/bee-world-500k_v3/round5/test_logs/long_good.txt")[1]
```

```python
vision = np.loadtxt("models/bee-world-500k_v3/round5/test_logs/mod_vision.txt")[1]
```

```python
smell = np.loadtxt("models/bee-world-500k_v3/round5/test_logs/long_smell.txt")[1]
```

```python
%matplotlib inline
```

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

sns.barplot(x=["Full senses", "Bad vision", "Bad smell"], y=[count_good(good), count_good(vision), count_good(smell)],ax=ax1)
ax1.set_title("Fraction of unsuccessful episodes")

ax = sns.violinplot(data=[filter_good(good), filter_good(vision), filter_good(smell)], cut=0, ax=ax2)
ax.set_xticklabels(["Full senses", "Bad vision", "Bad smell"])
ax2.set_title("Distribution of lengths for succesful episodes")
```

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

sns.barplot(x=["Full senses", "Bad vision", "Bad smell"], y=[count_good(good), count_good(vision), count_good(smell)],ax=ax1)
ax1.set_title("Fraction of unsuccessful episodes")

ax = sns.boxplot(data=[filter_good(good), filter_good(vision), filter_good(smell)], ax=ax2)
ax.set_xticklabels(["Full senses", "Bad vision", "Bad smell"])
ax2.set_title("Distribution of lengths for succesful episodes")
```
