```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python

SMALL_SIZE = 18
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
MAX_STEPS = 5000

def count_good(ar):
    return len(ar[ar==MAX_STEPS])/len(ar)

def filter_good(ar):
    return ar[ar!=MAX_STEPS]
```

```python
good = np.loadtxt("models/bee-world-500k_v3/round5/test_logs-one-wall/both-senses-no-noise.txt")[1]
vision = np.loadtxt("models/bee-world-500k_v3/round5/test_logs-one-wall/noisy-vision.txt")[1]
smell = np.loadtxt("models/bee-world-500k_v3/round5/test_logs-one-wall/noisy-smell.txt")[1]
both = np.loadtxt("models/bee-world-500k_v3/round5/test_logs-one-wall/both-senses-noisy.txt")[1]
```

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(15, 6))

sns.barplot(x=["Full senses", "Bad vision", "Bad smell", "No senses"], y=[count_good(good), count_good(vision), count_good(smell), count_good(both)], ax=ax1)
ax1.set_title("Fraction of unsuccessful episodes")
ax1.set_xticklabels(["Full senses", "Bad vision", "Bad smell", "No senses"], rotation=45)

ax1.set_ylabel("Failure rate [%]")

ax = sns.violinplot(data=[filter_good(good), filter_good(vision), filter_good(smell), filter_good(both)], cut=0, ax=ax2)
ax.set_xticklabels(["Full senses", "Bad vision", "Bad smell", "No senses"], rotation=45)
ax2.set_ylabel("Episode length")
ax2.set_title("Distribution of lengths for succesful episodes");

f.suptitle("Simple Environment");
```
